# FTSE 100 Stock Tracker — Ranking Model

## Objective

This notebook ranks FTSE 100 stocks based on return and risk metrics.

The goal is to identify stocks that offer the best risk-adjusted performance.

## Method

Stocks are ranked using:

- Total Return (higher is better)
- Volatility (lower is better)

We combine these into a simple ranking score.

```python
import pandas as pd

df = pd.read_csv("../data/processed/ftse100_metrics.csv")
df.head()
```

## Ranking logic

We rank stocks based on:

- High return = good
- Low volatility = good

We combine both into a single ranking score.

```python
df["return_rank"] = df["total_return"].rank(ascending=False)
```

```python
df["risk_rank"] = df["volatility"].rank(ascending=True)
```

```python
df["score"] = df["return_rank"] + df["risk_rank"]
```

```python
ranking = df.sort_values("score")
ranking.head(20)
```

```python
ranking.to_csv("../data/processed/ftse100_ranking.csv", index=False)
```

## Ranking complete

Stocks have been ranked based on total return and volatility.
A lower score indicates better risk-adjusted performance.

The findings from our list show a strong presence of banks and financial companies. Firms such as HSBC and M&G have benefited from higher interest rates and improved margins, which translates to stable and predictable cash flow. Our model rewards companies for delivering solid returns without excessive volatility.

Consumer staple brands are also predominatley seen in the top performers. Companies like Coca Cola and Tesco show good risk adjusted rankings. Their pricing power, stable earnings, and consistent demand allow them to rank highly.  

Overall, these rankings highlight a bigger picture, high growth sectors such as Tech and Oil are missing from our list. Our model appears to be favouring stable cash-flow businesses, ather than high variance sectors.
