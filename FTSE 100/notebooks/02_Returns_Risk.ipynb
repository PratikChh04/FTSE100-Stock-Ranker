# FTSE 100 Stock Tracker — Returns & Risk Analysis

## Objective

This notebook transforms the raw price data into performance metrics.

Using the adjusted closing prices collected in Notebook 01, we calculate:

- Total Returns
- Volatility (Risk)

These metrics form the foundation of the ranking model developed in Notebook 03.

## Method

We compute:

- Cumulative return for each stock over the selected time period
- Annualised volatility as a measure of price risk

These outputs allow us to compare performance consistency across FTSE 100 constituents before applying a ranking framework.

```python
import pandas as pd
import numpy as np

prices = pd.read_csv("../data/raw/ftse100_prices.csv")
prices.head()
```

```python
prices["date"] = pd.to_datetime(prices["date"])
prices = prices.set_index("date")
prices.head()
```

```python
returns = prices.pct_change()
returns.head()
```

```python
total_return = (prices.iloc[-1] / prices.iloc[0]) - 1
total_return.head()
```

```python
volatility = returns.std() * np.sqrt(252)
volatility.head()
```

```python
metrics = pd.DataFrame({
    "total_return": total_return,
    "volatility": volatility
})

metrics = metrics.dropna()
metrics.head()
```

```python
metrics.to_csv("../data/processed/ftse100_metrics.csv")
```

## Summary of Findings

We have successfully transformed raw price data into two key performance metrics:

- Total return, measuring overall price growth
- Volatility, serving as a proxy for risk

These return and risk metrics now provide the foundation for a structured ranking model.
