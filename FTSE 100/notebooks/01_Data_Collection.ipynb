# FTSE 100 Stock Tracker — Data Collection

## Objective
This notebook collects historical adjusted closing prices for FTSE 100 companies using the Yahoo Finance API.

The data will be used in subsequent notebooks to calculate returns, risk metrics, and build a stock ranking model.

## Download historical price data

We retrieve adjusted closing prices for all FTSE 100 tickers.

Adjusted Close is used because it accounts for:
- Dividends
- Stock splits
- Corporate actions

This ensures returns calculations are accurate.

```python
prices = yf.download(
    tickers,
    start="2023-01-01",
    auto_adjust=False,   # ensures "Adj Close" exists
    progress=False
)

adj_close = prices["Adj Close"]
adj_close = adj_close.reset_index().rename(columns={"Date": "date"})
```

```python
adj_close.to_csv("../data/raw/ftse100_prices.csv", index=False)
```
